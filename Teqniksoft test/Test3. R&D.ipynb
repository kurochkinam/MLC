```python
from bs4 import BeautifulSoup
import urllib.request
```

```python
def download(t_url):
    response = urllib.request.urlopen(t_url)
    data = response.read()
    txt_str = str(data)
    lines = txt_str.split("\\n")
    des_url = 'D:/Download/test.html'
    fx = open(des_url,"w")
    for line in lines:
        line=line.replace("\\t", "").replace("\\r", "")
        fx.write(line+ "\n")
    fx.close()

```

```python
url_name = "http://www.pcpao.org/general.php?strap=162906640000010420"
download(url_name)
```

```python
filename="D:/Download/test.html"
soup = BeautifulSoup(open(filename), "html.parser")
```

```python
print(soup.prettify())
```

```python
data = []
tables = soup.find_all('table')


for table in tables:
    #table_body = table.find('tbody')
    rows = table.find_all('tr')
    for row in rows:
        cols = row.find_all('td')
        cols = [ele.text.strip() for ele in cols]
        data.append([ele for ele in cols if ele]) 

```
