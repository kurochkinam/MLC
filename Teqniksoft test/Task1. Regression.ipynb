### Для dataset_task1.csv определить, как была смоделирована колонка y

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.neighbors import KNeighborsClassifier,KNeighborsRegressor
from sklearn.cluster import KMeans
from sklearn.preprocessing import  StandardScaler
from sklearn.utils import shuffle
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.linear_model import LinearRegression, Lasso, Ridge, SGDRegressor
from sklearn import metrics
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import cross_val_score, train_test_split, cross_validate, ShuffleSplit
```

```python
import seaborn as sns
real_features=["x_1", "x_2"]
cat_features=["x_3", "x_4"]
```

```python
# y distribution has a long tail on the right side, thus it is likely a mixture of several distributions
sns.distplot(data["y"], bins=50)
```

```python
# I applied kmeans clustering with 2, 3, 4 clusters. 2 & 3 seems to be a choice. 
#There are other methods of unsupervised clustering, bayesian, EM, etc, but I hadn't test them due to time limitations
y=np.array(data["y"])
clf2=KMeans(n_clusters=2, max_iter=500)
clf2.fit_predict(y.reshape(len(y), 1))
clf3=KMeans(n_clusters=3, max_iter=500)
clf3.fit_predict(y.reshape(len(y), 1))
clf4=KMeans(n_clusters=4, max_iter=500)
clf4.fit_predict(y.reshape(len(y), 1))
```

```python
plt.figure(figsize=(20,6))
plt.subplot(1,3,1)
plt.scatter(np.arange(0, len(y)) , y, c=clf2.labels_)
plt.title("2 clusters")
plt.subplot(1,3,2)
plt.scatter(np.arange(0, len(y)) , y, c=clf3.labels_)
plt.title("3 clusters")
plt.subplot(1,3,3)
plt.scatter(np.arange(0, len(y)) , y, c=clf4.labels_)
plt.title("4 clusters")
```

```python
y2_1, y2_0 =[], []
y=data["y"]
for i in range (0, len(y)):
    if clf2.labels_[i]==0:
        y2_0.append(y[i])
    else:
        y2_1.append(y[i])

y3_1, y3_0, y3_2 =[], [], []
y=data["y"]
for i in range (0, len(y)):
    if clf3.labels_[i]==0:
        y3_0.append(y[i])
    if clf3.labels_[i]==1:
        y3_1.append(y[i])
    else:
        y3_2.append(y[i])
```

```python
plt.figure(figsize=(20,6))
plt.subplot(2,3,1)
plt.hist(y3_0, bins=20)
plt.title("3 clusters, cluster 0")
plt.subplot(2,3,2)
plt.hist(y3_1,  bins=20)
plt.title("3 clusters, cluster 1")
plt.subplot(2,3,3)
plt.hist(y3_2,  bins=20)
plt.title("3 clusters, cluster 2")
plt.subplot(2,3,4)
plt.hist(y2_0, bins=20, color="r")
plt.title("2 clusters, cluster 0" )
plt.subplot(2,3,5)
plt.hist(y2_1,  bins=20, color="r")
plt.title("2 clusters, cluster 1")

```

### Regression

```python
# 2 nominal features, one uniformly distributed and one like looks like normal. As y is not normal itself, 
#linear models are performing poorly
sns.pairplot(data, kind="scatter")
```

```python
# normalized all real features and labels
scaler=StandardScaler()
scaled_data=scaler.fit_transform(data[real_features])
scaled_y=scaler.fit_transform(np.array(data["y"]).reshape((data["y"].shape[0], 1)))
scaled_data=pd.DataFrame(scaled_data, columns=['x_1', "x_2"])
```

```python
# processed categoric features
x_3=pd.get_dummies(data["x_3"], prefix="x_3")
x_4=pd.get_dummies(data["x_4"], prefix="x_4")
```

```python
data_all=pd.concat([scaled_data, x_3, x_4], axis=1)
```

```python
y=scaled_y
X=data_all

X_sh, y_sh=shuffle(X,y)
X_train, X_test, y_train, y_test=train_test_split(X_sh, y_sh, test_size=0.3)
```

```python

regression_result=pd.DataFrame(columns=["Regression",  "R2"])
regression_list=[ LinearRegression(),
                  Lasso(),
                  SGDRegressor(max_iter=1000),
                  Ridge(),
                  KNeighborsRegressor(),
                GaussianProcessRegressor()]


#scoring={
#        "MSE": metrics.mean_squared_error,
 #       "MAE": metrics.median_absolute_error}
for regression in regression_list:
    cv=cross_val_score(regression, X_train, y_train, cv=ShuffleSplit(test_size=0.3), scoring="r2", n_jobs=-1)
    entry=pd.DataFrame([[regression.__class__.__name__,  cv.mean()]],
                           columns=["Regression",  "R2"])
   
    regression_result=regression_result.append(entry, ignore_index=True)
regression_result=regression_result.sort_values("R2", ascending=False).reset_index(drop=True)#regression_result
```

```python
#this is cross validation results for different regression algorithms. GaussianProcessRegressor works in a best way
regression_result
```

```python
# Results on test sample
clf=GaussianProcessRegressor()
clf.fit(X_train, y_train)

print("R2 score {:.3}, MSE {:.3}".format (metrics.r2_score(y_test, clf.predict(X_test)), 
                                    metrics.mean_squared_error(y_test, clf.predict(X_test))))

```
