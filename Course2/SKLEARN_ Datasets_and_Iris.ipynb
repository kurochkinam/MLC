# Посчитать уникальные значения в np.array

```python
unique,counts= np.unique(Y_test, return_counts=True)
z=float(len(Y_test))
d=dict(zip(unique, counts))
```

```python
for key, value in d.items():
    print ("key {}: {}". format(key, round(value/z, 2)))
```

# KFOLD

```python
for train_indices, test_indices in cross_validation.KFold(10, n_folds=5, shuffle=True, random_state=42):
    print (train_indices, test_indices)
```

# Stratified Kfold

```python
myset=([0]*5+[1]*5)
print(myset)

for train_indices, test_indices in cross_validation.StratifiedKFold(myset, n_folds=2, shuffle=True, random_state=0):
    print(train_indices, test_indices)
```

# Shuffle & Stratified Shuffle 

```python
for train_ind, test_ind in cross_validation.ShuffleSplit(10, n_iter=8, test_size=0.2):
    print(train_ind, test_ind)
```

```python
myset=([0]*5+[1]*5)
for train_ind, test_ind in cross_validation.StratifiedShuffleSplit(myset, n_iter=4, test_size=0.2):
    print(train_ind, test_ind)
```

# Leave one out

```python
for train_ind, test_ind in cross_validation.LeaveOneOut(10):
    print(train_ind, test_ind)
```
