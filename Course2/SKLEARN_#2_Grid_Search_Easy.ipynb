```python
import numpy as np
import pandas as pd
from sklearn import datasets, linear_model, metrics, model_selection
```

```python
iris=datasets.load_iris()
```

```python
X_train, X_test, Y_train, Y_test=model_selection.train_test_split(iris.data, iris.target, test_size=0.3, random_state=1)
```

```python
classifier=linear_model.SGDClassifier(random_state=1, max_iter=100)
```

```python
classifier.get_params().keys()
```

# Подбор параметров "поиск по сетке". 

```python
# будет декартово произведение параметров, для каждого свое значение модели. Можно выбрать самое оптимальное. 
```

```python
parameters_grid={
    'alpha': np.linspace(0.0001, 0.001, 5),
    'loss':["hinge","log", "squared_hinge", "squared_loss" ],
    'penalty':['l1', "l2"],
    'n_iter': range(5,10)
}
```

```python
cv=model_selection.StratifiedShuffleSplit( test_size=0.2, n_splits=10, random_state=1)
cv1=cv.split(X_train, Y_train)
```

```python
grid_cv=model_selection.GridSearchCV(classifier, parameters_grid, scoring="accuracy", cv=cv1) # объект поиска по сетке
```

```python
%%time
grid_cv.fit(X_train, Y_train)
```

```python
grid_cv.best_estimator_
```

```python
print(grid_cv.best_score_)
```

```python
print(grid_cv.best_params_)
```

```python
grid_cv.grid_scores_[:10]
```

# Случайный поиск по сетке

```python
rand_grid_cv=model_selection.RandomizedSearchCV(classifier, parameters_grid, scoring="accuracy", cv=cv1, n_iter=20, random_state=1)
# объект рандомного поиска по сетке. n_iter задает число точек в сетке, которые будут просчитаны
```

```python
%%time
rand_grid_cv.fit(X_train, Y_train)
```

```python
rand_grid_cv.best_params_
```

```python
rand_grid_cv.best_score_
```
