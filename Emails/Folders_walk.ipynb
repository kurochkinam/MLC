```python

del_ = str.maketrans("","", string.punctuation)
del_galki=str.maketrans("","", "<>")
print("Z fhskf< > kjks!! ,dfd".translate(del_galki))
```

```python
#test_tree=os.walk("D:\Download\Trainingsmails_Anonym\Kino") 
#небуквенные знаки удалять?
def to_base(tree):
    base=pd.DataFrame(columns=["Text", "Label"])
    for address, _, file in tree: #проход по всем папкам-
        for ff in file: # проход по всем файлам в папке
            if not ff.endswith(".eml"):
                continue
            path = os.path.join(address,ff)
            with open(path, "r", errors='ignore', encoding="utf8") as f:
                text=[]
                for line in f:
                    if line.startswith("From:"):
                        line=line.replace("From: ", "").lower().translate(del_galki).replace("\n", "")
                        lines=line +" "
                    if line.startswith("Subject:"):
                        line=line.replace("Subject: ", "").lower().translate(del_).replace("\n", "").translate(del_galki)  #удалять переносы?
                        lines=lines + line
                        text.append(lines)            
            label=address.split("\\")[-1]
            row=pd.DataFrame({"Text":text, "Label": label})
            base=base.append(row, ignore_index=True)
    return base
```

```python
def flatten(l, ltypes=(list, tuple)):
    ltype = type(l)
    l = list(l)
    i = 0
    while i < len(l):
        while isinstance(l[i], ltypes):
            if not l[i]:
                l.pop(i)
                i -= 1
                break
            else:
                l[i:i + 1] = l[i]
        i += 1
    return ltype(l)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn import metrics
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
```

```python
X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=0.1, stratify=y)
```

```python
count_vectorizer = CountVectorizer()
counts = count_vectorizer.fit_transform(X_train)
len(count_vectorizer.get_feature_names())
```

```python
result=pd.DataFrame(columns=["max_features", "score"])
max_features= np.arange(1000,1300,25)

for i in max_features:
    
    pipeline = Pipeline([
        ('vectorizer',  CountVectorizer(max_features=i)),
        ('classifier',  MultinomialNB()) ])

    pipeline.fit(X_train, y_train)
    acc=metrics.accuracy_score(pipeline.predict(X_test) , y_test)
    row=pd.DataFrame([[i, acc ]],columns=["max_features", "score"] )
    #print(row)
    result=result.append(row, ignore_index=True)
result
```

```python
result1=pd.DataFrame(columns=["max", "min", "score"])
max_df= [200, 450, 390, 380]
min_df= [1, 2, 3]
for i in max_df:
    for j in min_df:
        pipeline = Pipeline([
            ('vectorizer',  CountVectorizer(max_df=i, min_df=j)),
            ('classifier',  MultinomialNB()) ])

        pipeline.fit(X_train, y_train)
        acc=metrics.accuracy_score(pipeline.predict(X_test) , y_test)
        row=pd.DataFrame([[i, j, acc ]],columns=["max", "min", "score"] )
        result1=result1.append(row, ignore_index=True)
result1
```

```python
result2=pd.DataFrame(columns=["n_grams", "score"])
max_features= [(1,2), (1,3)]

for i in max_features:
    
    pipeline = Pipeline([
        ('vectorizer',  CountVectorizer(ngram_range=(1,2))),
        ('classifier',  MultinomialNB()) ])

    pipeline.fit(X_train, y_train)
    acc=metrics.accuracy_score(pipeline.predict(X_test) , y_test)
    row=pd.DataFrame([[i, acc ]],columns=["n_grams", "score"] )
    #print(row)
    result2=result2.append(row, ignore_index=True)
result2
```
