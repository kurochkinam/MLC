# Assignment_Continuous_distribution


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts

%matplotlib inline
```

```python
distr=sts.chi2(5)
sample=distr.rvs(1000)
x=np.linspace(0,25,1000)
plt.hist(sample, bins=15, normed=True)
plt.plot(x, distr.pdf(x), 'r')
plt.xlabel("x")
plt.ylabel("F(x)/fractions")
plt.title("Chi2 density function & 1000 sample histogram")
```

# Выборки размером 5

```python
n1, n2, n3, n4=5 , 15, 30, 50
mean1=np.array([])
mean2=np.array([])
mean3=np.array([])
mean4=np.array([])
for i in range(1000):
    samp1=distr.rvs(n1)
    samp2=distr.rvs(n2)
    samp3=distr.rvs(n3)
    samp4=distr.rvs(n4)
    mean1=np.append(mean1, np.mean(samp1))
    mean2=np.append(mean2, np.mean(samp2))
    mean3=np.append(mean3, np.mean(samp3))
    mean4=np.append(mean4, np.mean(samp4))
```

```python
# У хи-2 распределения мат ожидание=N, а дисперсия =2n
EX=5
DX=2*5
x=np.linspace(0,10, 1000)
```

```python
fig=plt.figure()
fig, ax=plt.subplots(nrows=2, ncols=2, figsize=(8,8))
ax1, ax2, ax3, ax4 = ax.flatten()
ax1.hist(mean1, normed=True)
ax1.plot(x, sts.norm(loc=EX, scale=(DX/n1)**(1/2) ).pdf(x), 'r')
ax1.set_title("n=5")
ax2.hist(mean2, normed=True)
ax2.plot(x, sts.norm(loc=EX, scale=(DX/n2)**(1/2) ).pdf(x), 'r')
ax2.set_title("n=15")
ax3.hist(mean3, normed=True)
ax3.plot(x, sts.norm(loc=EX, scale=(DX/n3)**(1/2) ).pdf(x), 'r')
ax3.set_title("n=30")
ax4.hist(mean4, normed=True)
ax4.plot(x, sts.norm(loc=EX, scale=(DX/n4)**(1/2) ).pdf(x), 'r')
ax4.set_title("n=50")
```
